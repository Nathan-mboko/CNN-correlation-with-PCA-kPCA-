```python

import numpy as np
import matplotlib.pyplot as plt
```

```python
# Define the classes
classes = ['2S1', 'BRDM_2', 'BTR_60', 'D7', 'SLICY', 'ZIL131', 'ZSU_23_4']

# Input Pearson correlation values for each class (as an example, replace with your actual data)
pearson_corr_values = {
    '2S1': 0.32,
    'BRDM_2': 0.29,
    'BTR_60': 0.35,
    'D7': 0.30,
    'SLICY': 0.49,
    'ZIL131': 0.27,
    'ZSU_23_4': 0.5
}

# Extract Pearson correlation values from the dictionary
corr_values = [pearson_corr_values[cls] for cls in classes]

# Plot histogram
plt.figure(figsize=(10, 6))
plt.bar(classes, corr_values, color='skyblue')

# Add labels and title
plt.xlabel('Classes')
plt.ylabel('Pearson Correlation')
plt.title('Pearson Correlation for Various Classes')

# Display the values on top of the bars
for i, v in enumerate(corr_values):
    plt.text(i, v + 0.01, f"{v:.2f}", ha='center', fontweight='bold')

# Show the plot
plt.show()
```
